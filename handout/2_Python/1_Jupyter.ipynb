# How to interact with Jupyter

Jupyter allows one to combine chunks of text and chunks of code through a unified interface: the Jupyter notebook.  Each chunk is called a "cell".  Different kernels exist that allow for different programming languages to be used.

Execute the following cells using SHIFT + ENTER.

```python
# Here is a chunk of code
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
%matplotlib inline
```

Here's a plot:

```python
t = np.linspace(-pi, +pi)
y = np.sin(t)
plt.plot(t, y)
plt.show()
```
